```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
user = pd.read_csv('/content/drive/My Drive/users.csv')
post = pd.read_csv('/content/drive/My Drive/posts.csv')
views = pd.read_csv('/content/drive/My Drive/views.csv')
```

Importing Dataset

```
post = post.rename(columns={'_id':'post_id'})
```

```
post.describe()
```

This shows ctagory have some null values

```
post.fillna('project')
```

Filling Null values of category with name project 

```
user.head()
```

```
views.head()
```

```
user = user.rename(columns={'_id':'user_id'})
```

```
user['user_id'] = user.user_id.astype("category").cat.codes
views['user_id'] = views.user_id.astype("category").cat.codes
```

Mapping of user_id from 0 to 117

```
df = pd.merge(views,post,on='post_id')
```

Merging views and post on the basis of post_id

```
df.head(10)
```

```
no_of_views = df.groupby('category')['user_id'].count()
no_of_views = no_of_views(columns={'views'})
no_of_views.head(10)
```

Number of views got by a posts of a particular category


```
plt.hist(no_of_views,bins=100)
plt.show()
```

```
postmat = pd.DataFrame(index=user['user_id'],columns=df['category'].unique())
```

Creating a DtaFrame with indexes as user_id and column with category

```
postmat.head(10)
```

```
z=1
for column in postmat:
  y = postmat.columns.get_loc(column)
  for i in range(118):
    x =  df.loc[(df['category'] == column) & (df['user_id'] == i)]
    if x.empty:
      z=z+1
    else:
      postmat.iloc[i,y] = 1
    
```

Replace NaN by 1 if a user seen that particular post of that category


```
postmat.head(10)
```

Now we will take Photography as an example to see its correlation with postmat

```
photo_views = postmat['Photography']
photo_views.head()
```

```
similar_to_photo = postmat.corrwith(photo_views)
```

Calculating correlation

```
corr_photo = pd.DataFrame(similar_to_photo,columns=['Correlation'])
```

```
corr_photo.head()
```

For a better understanding we will add number of views also

```
corr_photo = corr_photo.join(no_of_views)
```

We would like to recommend posts which have some relatable number of views 

```
corr_photo[corr_photo['views']>5].sort_values('correlation',ascending=False)
```

Now for post of category Photography we can recommend 3 or 4 posts from this correlation table
